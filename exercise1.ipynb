```python
import pandas as pd
import matplotlib.pyplot as plt

data = {"Date": ["2024-10-01","2024-10-01","2024-10-02","2024-10-02","2024-10-03","2024-10-03"],
        "Product": ["A","B","A","B","A","B"], "Sales": [150, 200, 180, 220, 170, 210]}

df = pd.DataFrame(data)
df
```

```python
sum_a = df.groupby("Product")["Sales"].sum()["A"]
sum_b = df.groupby("Product")["Sales"].sum()["B"]

print("Total Sales for Product A: ", sum_a)
print("Total Sales for Product A: ", sum_b)
```

```python
A = df[df["Product"] == "A"]
B = df[df["Product"] == "B"]
A, B
```

```python
plt.figure(figsize = (8,8))
plt.plot(A["Date"], A["Sales"], marker="o", label = "A")
plt.plot(B["Date"], B["Sales"], color = "orange", marker = "o", label= "B")
plt.title("Daily Sales Trend by Product")
plt.legend(title = "Product")
plt.show()
```

```python
plt.figure(figsize = (8, 8))
plt.bar(["A", "B"],[sum_a,sum_b])
plt.title("Total Sales by Product")
plt.show()
```
