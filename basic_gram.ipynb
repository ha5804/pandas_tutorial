# 1.DataFrame

#### (1)create dataframe

```python
import pandas as pd

#make df from a dictionary

data = {"Name": ['Alice', 'Bob', 'Charlis'], "Age": [25, 32, 13], "city": ["Rome", "Milam", "seoul"]}
df = pd.DataFrame(data)
print(df.head())
print(df.columns) #return index column and dtype
```

#### (2) filter data

```python
df_filter = df[df["Age"] > 20]
#filter condition

print(df_filter)
```

#### (3) calculate condition with columns

```python
df_aggregate = df.groupby("city")['Age'].mean()
#도시별 평균 나이 계산
print(df_aggregate)
```

#### (4) create new columns

```python
df["Adult"] = df["Age"].apply(lambda x: "Yes" if x >= 18 else "No")
print(df)
```

#### (5) delete columns

```python
df_drop = df.drop(columns = ["Age"])
print(df_drop)
```

#### (6) save csv

```python
df_drop.to_csv("ex_pandas.csv", index = False)
```

# 2.Series

#### (1) make series

```python
ob = pd.Series(range(5))
print(ob)
```

#### (2)make series from a dictionary

```python
obj = pd.Series({"Jan" : 31, "Feb": 28, "Mar": 31})
print(obj)
```

#### (3)make series with scalar value

```python
a = pd.Series(10, index = range(0,3))
print(a)

b = pd.Series(15, index = range(1, 6, 2))
print(b)

c = pd.Series("Welcom to BBK", index = ["Hema", "Anup"])
print(c)
```
